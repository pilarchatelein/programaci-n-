1)1. Diseñe e implemente un programa en Python en donde se ingrese un número e informe:

a) si es par o impar.

b) si es múltiplo de 5.

c) si es múltiplo de 3.

d) si es múltiplo de 5 y 3 a la vez.



```python
n=int(input("Ingrese un numero: "))

if n%2==0 :
    print(n,"es un numero par")

if n%5==0:
    print(n,"es multiplo de 5")

if n%3==0:
    print(n,"es multiplo de 3")

if n%5==0 and n%3==0:
    print(n,"es multiplo de 5 y 3 a la vez")
```

Escriba un programa que pida tres números y diga si forman una sucesión aritmética o geométrica.
 Recuerda que:

En una sucesión aritmética la diferencia entre dos elementos consecutivos es siempre la misma y a esa
cantidad se le llama la diferencia de la sucesión aritmética.
En una sucesión geométrica el cociente entre dos elementos consecutivos es siempre el mismo y a esa
 cantidad se  le llama la razón de la sucesión geométrica.

```python
n1=int(input("ingrese un numero: "))
n2=int(input("ingrese un numero: "))
n3=int(input("ingrese un numero: "))
resta_1=n1-n2
resta_2=n2-n3

div1=n2/n1
div2=n3/n2
if resta_1==resta_2:
    print("es una sucesion aritmetica")

if div1==div2:
    print("es una sucesion geometrica")

else:
    print("no corresponde a ninguna sucesion")
```

Regular: Asistencia a clases 75%, notas de los parciales con promedio mayor o igual a 60% y ninguna de las menores a 50%.
Promocional: Asistencia a clases 75%, notas de los parciales con promedio mayor o igual a 75% y ninguna de las menores a 60%.
Pida al usuario todos los datos necesarios para establecer la condición y en el caso de promoción calcule la nota final en la asignatura si la

11)Durante una reunión con personal de guardia del Hospital, se planteó la necesidad de que el sistema clasifique la condición térmica de un paciente a partir de la temperatura axilar ingresada. Escriba un programa que permita cargar los datos y emitir la condición térmica correspondiente.

```python
temperatura_axilar=float(input("Ingrese la temperatura axilar de cada paciente: "))
if temperatura_axilar<35.0:
  print("Condición: Hipotermia")

elif 35.0<temperatura_axilar<37.4:
  print("Condición: Normal")
elif 37.5<temperatura_axilar<37.9:
  print("Condición: Febrícula")
elif 38.0<temperatura_axilar<39.9:
  print("Condición: Fiebre")
elif temperatura_axilar>40.0:
  print("Condición: Hipertermia")
```
